```python
import requests
from bs4 import BeautifulSoup

url = "https://www.nextage.jp/stockcar/"
headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }

response = requests.get(url, headers=headers)
if response.status_code != 200:
    print("Could not get page information", response.status_code)

soup = BeautifulSoup(response.text, "html.parser")
```

```python
car_elements = soup.find_all("div", class_="s_k_sbox pi_imp")
car_list = []
for car in car_elements:
        try:
            # 車名
            name = car.find("span", itemprop="name").text.strip() if car.find("span", itemprop="name") else "不明"
            # 支払総額
            total_price = car.find("span", class_="red").text.strip() if car.find("span", class_="red") else "不明"
            
            # テーブル内の詳細情報
            table = car.find("table", class_="spec")
            year = mileage = inspection = "不明"

            if table:
                rows = table.find_all("tr")
                for row in rows:
                    th = row.find("th").text.strip()
                    td = row.find("td").text.strip()
                    if th == "年式":
                        year = td
                    elif th == "走行":
                        mileage = td
                    elif th == "車検":
                        inspection = td

            car_list.append({
                "name": name,
                "total_price": total_price,
                "year": year,
                "mileage": mileage,
                "inspection": inspection
            })
        except AttributeError:
            continue
```

```python
for car in car_list:
        print(f"車名: {car['name']}")
        print(f"支払総額: {car['total_price']}万円")
        print(f"年式: {car['year']}")
        print(f"走行距離: {car['mileage']}")
        print(f"車検: {car['inspection']}")
        print("-" * 50)
```
